```python
# I wanted to comapare GoogleTrends data on searches for pornography in the Netherlands with the CBS data. Here is the installation fo the package, followed by the defintions of the results I want (hl='en-NL' pulling up data from the Netherlands in English, tz=300 indicating the timezone.)

import pandas as pd
import pytrends
from pytrends.request import TrendReq
!pip install pytrends
pytrend = TrendReq(hl='en-NL', tz=300)
```

```python
# Here I am defining the keywords I want to be searched, as well as the time period. These create a dataset using the build_payload function to show 'interest over time'.

keywords = ['porn', 'pornography','porno']
pytrend.build_payload(kw_list=keywords, timeframe='2014-01-01 2014-12-31')
interest_over_time_df_2014=pytrend.interest_over_time()

interest_over_time_df_2014
```

```python
# I have then asked for describtive statistics on this data to show the mean interest over the year 2014.
interest_over_time_df_2014.describe()
```

```python
# I have created the same dataset for the year 2022.

keywords = ['porn', 'pornography', 'porno']
pytrend.build_payload(kw_list=keywords, timeframe='2022-01-01 2022-12-31')
interest_over_time_df_2022=pytrend.interest_over_time()

interest_over_time_df_2022
```

```python
interest_over_time_df_2022.describe()
```
